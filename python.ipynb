```python
import turtle

# Set up the screen
screen = turtle.Screen()
screen.bgcolor("lightblue")
screen.title("Pushkar Ki Himanshi")

# Create a turtle object
pen = turtle.Turtle()
pen.speed(1)
pen.color("darkblue")

# Function to write text
def write_text(text, x, y, font_size=30, font_style="bold"):
    pen.penup()
    pen.goto(x, y)
    pen.pendown()
    pen.write(text, align="center", font=("Arial", font_size, font_style))

# Write "Pushkar Ki Himanshi"
write_text("Pushkar", -100, 50, font_size=40)
write_text("Ki", 0, 0, font_size=30)
write_text("Himanshi", 100, -50, font_size=40)

# Hide the turtle
pen.hideturtle()

# Keep the window open
turtle.done()
```

```python
def is_vowel(char):
    """Checks if a given character is a vowel."""
    vowels = "aeiouAEIOU"
    return char in vowels

# Get input from the user
char = input("Enter a character: ")

# Check if the input is a single character
if len(char) == 1:
    if is_vowel(char):
        print(f"'{char}' is a vowel.")
    else:
        print(f"'{char}' is not a vowel.")
else:
    print("Please enter a single character.")
```
